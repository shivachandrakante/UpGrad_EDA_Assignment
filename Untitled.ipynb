```python
df['revol_util_range']=df['revol_util'].apply(lambda x:float(x.strip('%')))
```

```python
def revol_range(column):
    if(column<=10):
        return 1
    elif(column>10 and column<=20):
        return 2
    elif(column>20 and column<=30):
        return 3
    elif(column>30 and column<=40):
        return 4
    elif(column>40 and column<=50):
        return 5
    elif(column>50 and column<=60):
        return 6
    elif(column>60 and column<=70):
        return 7
    elif(column>70 and column<=80):
        return 8
    elif(column>80 and column<=90):
        return 9
    elif(column>90 and column<=100):
        return 10
    
    
    
df['revol_util_range']=df['revol_util_range'].apply(revol_range)
```
